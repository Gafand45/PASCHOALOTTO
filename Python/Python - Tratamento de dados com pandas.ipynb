Resumindo dados pela coluna 'Status' e 'Ano'

```python
import pandas as pd

csv_tele = pd.read_csv('./telemarketing_atendimento.csv', sep=',')

#Substituindo da coluna avaliacao os valores nulos por 0
csv_tele['Avaliacao'] = csv_tele['Avaliacao'].fillna(0)

#Transformando a coluna Data_Atendimento em data reconhecida pelo python
csv_tele['Data_Atendimento'] = pd.to_datetime(csv_tele['Data_Atendimento'])
csv_tele['Ano'] = csv_tele['Data_Atendimento'].dt.year

filtro = csv_tele.groupby(['Status','Ano']).size().reset_index(name='Quantidade')

display(filtro)
```

Resumindo dados pela coluna 'Status' , 'Ano' e 'Mes

```python
import pandas as pd

csv_tele = pd.read_csv('./telemarketing_atendimento.csv', sep=',')

#Substituindo da coluna avaliacao os valores nulos por 0
csv_tele['Avaliacao'] = csv_tele['Avaliacao'].fillna(0)

#Transformando a coluna Data_Atendimento em data reconhecida pelo python
csv_tele['Data_Atendimento'] = pd.to_datetime(csv_tele['Data_Atendimento'])
csv_tele['Ano'] = csv_tele['Data_Atendimento'].dt.year
csv_tele['Mes'] = csv_tele['Data_Atendimento'].dt.month_name()

filtro = csv_tele.groupby(['Status','Ano','Mes']).size().reset_index(name='Quantidade')

display(filtro)
```
